```python
# 6. Часто для качественного анализа аудитории использую подходы, 
# основанные на сегментации. 
# Используя python, построй RFM-сегментацию пользователей, 
# чтобы качественно оценить свою аудиторию. 
# В кластеризации можешь выбрать следующие метрики: 
# R - время от последней покупки пользователя до текущей даты, 
# F - суммарное количество покупок у пользователя за всё время, 
# M - сумма покупок за всё время. Подробно опиши, как ты создавал кластеры. 
# Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. 
# Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2) 
# имеет границы метрик recency от 130 до 500 дней, 
# frequency от 2 до 5 заказов в неделю, 
# monetary от 1780 до 3560 рублей в неделю.
```

```python
#Используем прошлый дф из другого задания, где смерджены все 3 дф
c_o_i
```

```python
# Преобразование столбца с датой покупки в тип datetime
c_o_i['order_purchase_timestamp'] = pd.to_datetime(c_o_i['order_purchase_timestamp'])
```

```python
# Определение текущей даты
current_date = pd.to_datetime('2023-05-22')
```

```python
# Создание RFM-метрик
rfm_data = c_o_i.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (current_date - x.max()).days,  # Recency
    'order_id': 'count',  # Frequency
    'price': 'sum'  # Monetary
})
rfm_data.columns = ['Recency', 'Frequency', 'Monetary']
```

```python
rfm_data.head()
```

```python
# Масштабирование RFM-метрик
rfm_scaled = (rfm_data - rfm_data.mean()) / rfm_data.std()
```

```python
# Кластеризация методом K-средних
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(rfm_scaled)
```

```python
# Присвоение меток кластеров пользователям
rfm_data['Cluster'] = kmeans.labels_
```

```python
# Определение границ метрик для интерпретации
recency_bins = [0, 30, 60, 90, 120, float('inf')]
frequency_bins = [1, 2, 3, 4, 5, float('inf')]
monetary_bins = [0, 1000, 2000, 3000, 4000, float('inf')]
```

```python
# Описание RFM-сегментов c указанием границ
rfm_data['RecencyGroup'] = pd.cut(rfm_data['Recency'], bins=recency_bins, labels=False)
rfm_data['FrequencyGroup'] = pd.cut(rfm_data['Frequency'], bins=frequency_bins, labels=False)
rfm_data['MonetaryGroup'] = pd.cut(rfm_data['Monetary'], bins=monetary_bins, labels=False)
```

```python
segment_descriptions = rfm_data.groupby(['RecencyGroup', 'FrequencyGroup', 'MonetaryGroup']).agg({
    'Recency': ['min', 'max'],
    'Frequency': ['min', 'max'],
    'Monetary': ['min', 'max']
})
segment_descriptions.columns = ['RecencyMin', 'RecencyMax', 'FrequencyMin', 'FrequencyMax', 'MonetaryMin', 'MonetaryMax']
segment_descriptions.reset_index(inplace=True)
```

```python
# Вывод описания RFM-сегментов
for i, segment in segment_descriptions.iterrows():
    r_desc = f"recency от {segment['RecencyMin']} до {segment['RecencyMax']} дней"
    f_desc = f"frequency от {segment['FrequencyMin']} до {segment['FrequencyMax']} заказов"
    m_desc = f"monetary от {segment['MonetaryMin']} до {segment['MonetaryMax']} рублей"
    print(f"RFM-сегмент {segment['RecencyGroup']}{segment['FrequencyGroup']}{segment['MonetaryGroup']} ({r_desc}, {f_desc}, {m_desc})")
```
